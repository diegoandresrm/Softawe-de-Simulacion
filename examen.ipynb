#### 1. Se solicita al usuario ingresar por teclado 10 valores numéricos, posteriormente guardar estos valores en una lista. Crear una función que tenga como argumentos la lista mencionada, y devuelva: La lista ordenada de mayor a menor, el promedio y el valor máximo. Llamar a la función e imprimir el resultado. No utilizar numpy para esta operación

```python

def val_numericos(numeros):
    numeros.sort(reverse=True)
    numeros_ordenados = numeros

    suma = sum(numeros)
    cantidad = len(numeros)
    promedio = suma / cantidad

    valmaximo = numeros[0]

    return numeros_ordenados, promedio, valmaximo

valores = []
contador = 0

print("Ingresar 10 números enteros.")

while contador < 10:
    try:
        valor_ingresado = int(input("Ingresa el valor: "))
        valores.append(valor_ingresado)
        contador = contador + 1
    except :
        print("Se debes ingresar un número valido")

lista_final, promedio_final, maximo_final = val_numericos(valores)

print("Lista Ordenada :", lista_final)
print("Promedio de los valores:", promedio_final)
print("Valor Máximo:", maximo_final)
```

#### 3.  Generar una gráfica dinámica donde se visualizará una función seno con ruido randómico de amplitud 0.2. Para esto se puede:
#### - generar datos distribuidos linealmente de 0 a 10 para el eje X
#### - generar los datos para el eje  Y=sin⁡(X) + ruido
#### - usar una función para realizar plot dinámico que reciba el dato X,Y


```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output

Duracion = 10.0         
Amplitud = 0.2   
Tiempo = 0.02            


def generar_seno_con_ruido(x):
   
    ruido = np.random.uniform(-Amplitud, Amplitud)
    return np.sin(x) + ruido


xs = []
ys = []

def plotear_valores(x, y):
    xs.append(x)
    ys.append(y)

    clear_output(wait=True)            
    plt.figure(figsize=(8,4))
    plt.plot(xs, ys, '-', linewidth=1, label='sin(x)')
    plt.scatter(xs, ys, s=10)          
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Seno con ruido')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    t_inicio = time.time()
    elapsed = 0.0

    while elapsed < Duracion:
        elapsed = time.time() - t_inicio            
        X = (elapsed / Duracion) * 10.0
        Y = generar_seno_con_ruido(X)

        plotear_valores(X, Y)

        time.sleep(Tiempo)

    clear_output(wait=True)
    plt.figure(figsize=(8,4))
    plt.plot(xs, ys, '-', linewidth=1, label='sin(x) + ruido')
    plt.scatter(xs, ys, s=10)
    plt.xlabel('X')
    plt.ylabel('Y = sin(X) + ruido')
    plt.title('Gráfica')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

- Generamos la funcion con ruido 

"np.random.uniform(-Amplitud, Amplitud)" genera un número aleatorio dentro del rango de -0.2 y +0.2.

"np.sin(x)" calcula el valor de la función seno para el valor actual de x.

"sin(x) + ruido" suma a la funcion con el ruido
