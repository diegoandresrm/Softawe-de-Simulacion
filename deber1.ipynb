**1. Serie de Fibonacci**

Escribe un programa que imprima en pantalla los n primeros números de la Serie de Fibonacci, donde n es ingresado por teclado.

Nota: La serie de Fibonacci comienza con 0 y 1; cada número siguiente se obtiene como la suma de los dos anteriores.

1. Pedimos el numero al usuario 

```python
entrada_n = input("Ingresa la cantidad de numeros que desea (N): ")
N = int(entrada_n)
```

2. Inizializamos los primeros dos terminos
```pyhton
a = 0  # Primer termino (F0)
b = 1  # Segundo termino (F1)
contador = 0 # Contador para saber cuantos numeros hemos impreso
```

3. Generamos un bucle, mientras el contador sea menor que N, repite todo lo que esta dentro 
``` python
while contador < N:
```
4. Imprimer el numero actual 
``` python 
   print(a, end=" ")
```
5. Calcula el siguiente, ya que el siguiente numero es la suma de los dos anteriores
```python
    siguiente_termino = a + b
```
6. El valor que estaba en "b" se mueve a "a"
```pyhton 
a = b
```
7. Actualiza el valor, el numero recien calculado se mueve a b. Este se convierte en el "segundo numero" para la siguiente suma
```python 
  b = siguiente_termino
```
8. Aumenta la cuenta
```python
    contador = contador + 1
```



````python
entrada_n = input("Ingresa la cantidad de numeros que desea (N): ")
N = int(entrada_n)

a = 0  
b = 1  
contador = 0 

print(f"Los primeros {N} numeros de Fibonacci son:")


while contador < N:
    
    print(a, end=" ")
    siguiente_termino = a + b
    a = b
    b = siguiente_termino
    
    contador = contador + 1
````

Escribe un programa que imprima en pantalla la tabla de multiplicar de n del 1 al 12, donde n es ingresado por teclado.

1. Función que nos va a solicitar el numero del cual queremos la tabla 
```python
n = int(input("Ingresa el número para la tabla de multiplicar (n): "))
```

2. Imprimimos el encabezado la "n" será reemplazada por el número que nosotros le pidamos al programa

3. Hacer un bucle para repetir la acción, significa que producira números del 1 al 12.El for es la instrucción que iniciará el bucle, y la i es la variable que usamos para guardar cada número de la secuencia

````python
n = int(input("Ingresa el número para la tabla de multiplicar (n): "))

print(f"\nTabla del {n}:")

for i in range(1, 13):
    print(f"{n} x {i} = {n * i}")
````

```python
import random
```

- Si el ultimo intento no fuel el numero secreto y al mismo tiempo ya alcanzo el maximo de intentos (3), entonces mostrara el mensaje que el jugador perdio


````python
import random
numero_secreto = random.randint(1, 10)
intentos = 0 
max_intentos = 3

print("JUEGO DE ADIVINANZA")
print(f"Tienes {max_intentos} intentos para adivinar un número entre 1 y 10.")

while intentos < max_intentos:
    
    try:
        adivinanza = int(input(f"\nIntento #{intentos + 1}: ¿Cuál es tu número? "))
    except ValueError:
        print("El programa solo acepta número, intenta de nuevo")
        continue 

    intentos += 1
    
    if adivinanza == numero_secreto:
        print(f"\n Felicidades ganaste {numero_secreto}.")
        break 
        
    elif adivinanza < numero_secreto:
        print("El número secreto es mayor.")
        
    else:
        print("El número secreto es menor.")


if adivinanza != numero_secreto and intentos == max_intentos:
    print(f"Perdiste el número secreto era {numero_secreto}.")
````

**4. Año bisiesto**

Escribe un programa que reciba un año como entrada y determine si es bisiesto o no.

Pista: Un año es bisiesto si es divisible por 4, pero no por 100, excepto si también es divisible por 400.

1. Pedimos el año y almacenamos 
- Solicitamos el año
```python
entrada_año = input("Ingresa el año: ")
```
- Covertimos a entero 
```python
año = int(entrada_año) 
```

2. Evaluamos si el año es bisisesto, alcamenamos el resultado y realizamos dos condiciones
```python
es_bisiesto = (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0)
```

3. Mostramos el resultado, si el año es bisiesto o si el año no es bisiesto
```python
if es_bisiesto:
    print(f"El año {año} Es un año bisiesto.")
else:
    print(f"El año {año} No es un año bisiesto.")
```


````python
entrada_año = input("Ingresa el año: ")
año = int(entrada_año) 

es_bisiesto = (año % 4 == 0 and año % 100 != 0) or (año % 400 == 0)

if es_bisiesto:
    print(f"El año {año} Es un año bisiesto.")
else:
    print(f"El año {año} No es un año bisiesto.")
````

**5. Conteo de números impares**

Escribe un programa que pida al usuario un número positivo N, cuente cuántos números impares hay entre 1 y N, y los muestre por pantalla.

1. Definimos una función
```python
entrada_n = input("Ingresa un numero positivo (N): ")
```
2. Convertimos a entero
```python
N = int(entrada_n)
```
3. Creamos un contador donde se iran sumando los numeros impares que nos pide el programa
```python 
contador_impares = 0
```
4. Iniciamos el bucle, donde por cada numro en la secuencia desde el 1 hasta el numero que ingresemos asigna el valor a la variable numero
```python
for numero in range(1, N + 1):
```
5. Calculamos el residuo de una division 
```python
 if numero % 2 != 0:
```
6. Esta linea se ejecuta si el numero es impar y aumenta su valor en una unidad 
```python
  contador_impares = contador_impares + 1
```
7. Muestra el resultado 
```python
print(f"Hay {contador_impares} numeros impares entre 1 y {N}.")
```

````python
entrada_n = input("Ingresa un numero positivo (N): ")
N = int(entrada_n)

contador_impares = 0

for numero in range(1, N + 1):

    if numero % 2 != 0:
        contador_impares = contador_impares + 1
        
print(f"Hay {contador_impares} numeros impares entre 1 y {N}.")
````
